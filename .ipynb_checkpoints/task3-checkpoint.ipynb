```python
#Task 1 Basic text processing pipeline
import nltk
texts='''Education is the process of facilitating learning, or the acquisition of knowledge, skills, values, beliefs, and habits. Educational methods include teaching, training, storytelling, discussion and directed research. Education frequently takes place under the guidance of educators, however learners can also educate themselves. Education can take place in formal or informal settings and any experience that has a formative effect on the way one thinks, feels, or acts may be considered educational. The methodology of teaching is called pedagogy.'''

for text in texts:
    sentences=nltk.sent_tokenize(texts)
    for sentence in sentences:
        words=nltk.word_tokenize(sentence)
        tagged=nltk.pos_tag(words)
        print(tagged)
```
