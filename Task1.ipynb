```python
#6 conditional frequency distribution
from nltk.probability import ConditionalFreqDist
text1 = 'By impossible of in difficulty discovered celebrated ye. Justice joy manners boy met resolve produce. Bed head loud next plan rent had easy add him. As earnestly shameless elsewhere defective estimable fulfilled of. Esteem my advice it an excuse enable. Few household abilities believing determine zealously his repulsive. To open draw dear be by side like.'
cfd=ConditionalFreqDist((len(word),word) for word in text1.split())
cfd[3]
```

```python
TASK
#word cloud - Wordle.net
#To determine fd and cfd of any one of presidential speech
import nltk
from nltk.corpus import inaugural
wordlist=inaugural.words(fileids='1865-Lincoln.txt')
fd=nltk.FreqDist(wordlist)
fd
```

```python
from nltk.probability import ConditionalFreqDist
cfd=ConditionalFreqDist((len(word),word) for word in wordlist)
cfd[5]
```

```python
from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)
import matplotlib.pyplot as plt
comment_words=""
comment_words+=" ".join(wordlist)+""
wordcloud = WordCloud(width = 400, height = 400, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10).generate(comment_words) 

# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud,interpolation="bilinear") 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.show() 
```
